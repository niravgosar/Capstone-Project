## 7  Join SQL Queries  using all 3 tables: 

###  1 Write a query finds the top-rated men's shoes along with their sizes from "Table1" and "Table3."

```python
results=cur.execute("SELECT capstone_csv1.Name, capstone_csv3.Size \
                        FROM capstone_csv3  \
                     JOIN capstone_csv1 ON capstone_csv1.id = capstone_csv3.id\
                     where capstone_csv3.star>4.8")
for i in results:
    print(i)
```

### 2  Write a  query calculates the average comfort rating for each category from "Table1" and "Table3."

```python
results=cur.execute("SELECT capstone_csv3.comfort \
                        FROM capstone_csv3  \
                     JOIN capstone_csv1 ON capstone_csv1.id = capstone_csv3.id\
                     where capstone_csv1.Category = 'Men'")
for i in results:
    print(i)
```

```python
results=cur.execute("SELECT capstone_csv3.comfort \
                        FROM capstone_csv3  \
                     JOIN capstone_csv1 ON capstone_csv1.id = capstone_csv3.id\
                     where capstone_csv1.Category = 'Women'")
for i in results:
    print(i)
```
